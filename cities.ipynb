```python
import numpy as np
import pandas as pd
```

```python
simplemaps = pd.read_csv(r'worldcities.csv', on_bad_lines='skip')
simplemaps.head()
```

```python
# get list with column names
list(simplemaps.columns)
```

```python
# drop unnecessary columns
to_drop=['city_ascii', 'lat', 'lng', 'iso2', 'iso3', 'capital', 'population', 'id']
simplemaps.drop(to_drop, inplace=True, axis=1)
simplemaps.head(10)
```

```python
# export as csv file
simplemaps.to_csv(r'cities.csv', index=False);
```

```python
# export to Excel file
read_file = pd.read_csv (r'cities.csv')
read_file.to_excel (r'cities.xlsx', index = None, header=True)
```
